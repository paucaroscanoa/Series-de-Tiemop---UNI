```python
import numpy as np
import matplotlib.pyplot as plt

# Configuracin del nmero de pasos
n_steps = 1000

# Inicializacin del random walk con un punto de partida de cero
random_walk = [0]

# Generacin del random walk
for _ in range(1, n_steps):
    step = random_walk[-1] + np.random.normal()
    random_walk.append(step)

# Visualizacin del random walk
plt.figure(figsize=(10, 6))
plt.plot(random_walk, label="Random Walk", color="blue")
plt.title("Ejemplo de Random Walk")
plt.xlabel("Nmero de pasos")
plt.ylabel("Valor del Random Walk")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros para el ruido blanco
n = 100  # Nmero de puntos
media = 0  # Media
desviacion_estandar = 1  # Desviacin estndar

# Generacin del ruido blanco
ruido_blanco = np.random.normal(media, desviacion_estandar, n)

# Graficar el ruido blanco
plt.figure(figsize=(10, 6))
plt.plot(ruido_blanco, color="blue", label="Ruido Blanco")
plt.axhline(y=0, color='gray', linestyle='--')
plt.title("Ejemplo de Ruido Blanco")
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.legend()
plt.show()
```
